### Preprocessing

##### Normalize the distribution of data

```python
X = X.apply(lambda col : np.log1p(col))
```

##### Encoding catigorical data

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
y = label_encoder.fit_transform(df['class'])
y = pd.DataFrame({'class':y})
```

```python
df = pd.concat([X,y],axis=1)
```

```python
df['class'].value_counts()
```

##### Drop the outliers

```python
for col in X.columns:
    Q1 = df[col].quantile(.25)
    Q3 = df[col].quantile(.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
```

```python
df = df.reset_index().drop(columns=['index'],axis=1)
```

#### Scaling Data

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
df[X.columns] = scaler.fit_transform(df[X.columns])
```

#### Drop duplicated data

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

#### Balance the data

```python
X = df.drop(columns=['class'],axis=1)
y = df['class']
```

```python
from imblearn.over_sampling import SMOTE
```

```python
smote=SMOTE(random_state=42) 
X,y=smote.fit_resample(X,y)
```

```python
y.value_counts()
```

#### Splitting Data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

## Modeling and Evaluation

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, recall_score, f1_score, classification_report, roc_auc_score, roc_curve
```

#### 1. Multinomial Logistic Regression

```python
M = LogisticRegression(multi_class='multinomial',penalty=None,solver='saga')
M.fit(X_train, y_train)
```

```python
y_pred_M = M.predict(X_test)
```

```python
print(classification_report(y_test, y_pred_M, target_names=label_encoder.classes_))
```

```python
cm_M = confusion_matrix(y_test, y_pred_M)
sns.heatmap(cm_M, annot=True, cmap='Blues', fmt='d')
```

#### 2. Multinomial Lasso Logistic Regression

```python
M_l1 = LogisticRegression(multi_class='multinomial',penalty='l1',solver='saga')
M_l1.fit(X_train, y_train)
```

```python
y_pred_Ml1 = M_l1.predict(X_test)
```

```python
print(classification_report(y_test, y_pred_Ml1, target_names=label_encoder.classes_))
```

```python
cm_Ml1 = confusion_matrix(y_test, y_pred_Ml1)
sns.heatmap(cm_Ml1, annot=True, cmap='Blues', fmt='d')
```

#### 3. Multinomial Ridge Logistic Regression

```python
M_l2 = LogisticRegression(multi_class='multinomial',penalty='l2',solver='saga')
M_l2.fit(X_train, y_train)
```

```python
y_pred_Ml2 = M_l2.predict(X_test)
```

```python
print(classification_report(y_test, y_pred_Ml1, target_names=label_encoder.classes_))
```

```python
cm_Ml2 = confusion_matrix(y_test, y_pred_Ml2)
sns.heatmap(cm_Ml2, annot=True, cmap='Blues', fmt='d')
```

#### 4. Multinomial ElasticNet Logistic Regression

```python
M_EN = LogisticRegression(multi_class='multinomial',penalty='elasticnet',solver='saga',l1_ratio=0.5)
M_EN.fit(X_train, y_train)
```

```python
y_pred_MEN = M_EN.predict(X_test)
```

```python
print(classification_report(y_test, y_pred_MEN, target_names=label_encoder.classes_))
```

```python
cm_MEN = confusion_matrix(y_test, y_pred_MEN)
sns.heatmap(cm_MEN, annot=True, cmap='Blues', fmt='d')
```

#### 5. One vs Rest Logistic Regression

```python
OVR = LogisticRegression(multi_class='ovr',solver='liblinear')
OVR.fit(X_train, y_train)
```

```python
y_pred_OVR = OVR.predict(X_test)
```

```python
print(classification_report(y_test, y_pred_OVR, target_names=label_encoder.classes_))
```

```python
cm_OVR = confusion_matrix(y_test, y_pred_OVR)
sns.heatmap(cm_OVR, annot=True, cmap='Blues', fmt='d')
```

### Model Comparison and Interpretation

Multinomial strategy generally performed better than One_vs_Rest for this food classification task, as evidenced by higher accuracy and F1 scores.

#### How regularization affected model performance and generalization?

In this case regularization didn't make any improvment in accuracy or f1 score. \
The regularized model get worse test metrics than multinomial without regularization.

#### How data balancing methods improved classification results. 

Data balancing was the most thing made a good perfomance improvment .\
It led to more uniform precision and recall across all food types.

The best performing model is the multibomial logistic regression without regularization .\
